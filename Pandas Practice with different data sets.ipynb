```python
#Which was the most ordered item? and How many items were ordered?
```

```python
C= A.groupby('item_name')
C= C.sum()
C= C.sort_values(['quantity'], ascending = False)
C.head()
```

```python
#What was the most ordered item in the choice_description column?
```

```python
D=A.groupby('choice_description').sum()
D=D.sort_values(['quantity'], ascending=False)
D.head()

```

```python
#Turn the item price into a float
```

```python
dollar = lambda x: float(x[1:-1])
A.item_price = A.item_price.apply(dollar)
A.item_price
```

```python
#How much was the revenue for the period in the dataset?
```

```python
# this is for single quantity 
A['item_price'].sum()
```

```python
#this is for multiple quantity
revenue =(A['item_price'] * A['quantity']).sum()
revenue
```

```python
crime.head()
```

```python
#Convert the type of the column Year to datetime64
```

```python
crime.info()
#here year is int64
```

```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

```python
#Set the Year column as the index of the dataframe
```

```python
crime.index
```

```python
crime = crime.set_index('Year')
```

```python
crime.head()
```

```python
#Group the year by decades and sum the values
```

```python
crimes = crime.resample('10AS').sum()

population1 = crime['Population'].resample('10AS').max()

crimes['Population'] = population1

crimes
```
